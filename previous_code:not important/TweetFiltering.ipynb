```python
with open("tweets.txt",'r') as f:
    get_all=f.readlines()
with open("tweets2.txt",'w') as f:
    for i,line in enumerate(get_all,1):         ## STARTS THE NUMBERING FROM 1 (by default it begins with 0)    
        if '{"created_at":' in line and i != 1:                              ## OVERWRITES line:2
            f.writelines("," +line)
        elif i == 1:
            f.writelines("[" +line)
        else:
            f.writelines(line)
    f.writelines("]")
```
